```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")
cur = conn.cursor()

cur.execute("CREATE TABLE customers (customer_id INT, name TEXT, segment TEXT);")
cur.execute("CREATE TABLE orders (order_id INT, customer_id INT, amount REAL, order_date TEXT);")

customers = [(101, "Alpha Co", "Enterprise"), (102, "Beta LLC", "SMB"), (103, "Gamma Inc", "Enterprise"), (104, "Delta Corp", "SMB"), (105, "Epsilon Ltd", "Consumer")]
orders = [(1, 101, 1200.0, "2026-01-10"), (2, 101, 800.0, "2026-02-15"), (3, 102, 350.0, "2026-01-20"), (4, 103, 2500.0, "2026-01-25"), (5, 103, 1500.0, "2026-03-01"), (6, 104, 450.0, "2026-02-10"), (7, 105, 150.0, "2026-02-28"), (8, 102, 200.0, "2026-03-12")]

cur.executemany("INSERT INTO customers VALUES (?,?,?);", customers)
cur.executemany("INSERT INTO orders VALUES (?,?,?,?);", orders)
conn.commit()

query = "SELECT c.segment AS Customer_Segment, COUNT(DISTINCT c.customer_id) AS Total_Customers, COUNT(o.order_id) AS Total_Orders, SUM(o.amount) AS Total_Revenue, ROUND(AVG(o.amount), 2) AS Avg_Order_Value, ROUND(SUM(o.amount) / COUNT(DISTINCT c.customer_id), 2) AS Customer_LTV FROM customers c JOIN orders o ON c.customer_id = o.customer_id GROUP BY c.segment ORDER BY Total_Revenue DESC;"

df_sql = pd.read_sql_query(query, conn)
print("--- SQL FINANCIAL & LTV ANALYTICS ---")
print(df_sql)

conn.close()
```

```python
df_sql.to_csv("sql_ltv_customer_analytics.csv", index=False)
print("Data exported to sql_ltv_customer_analytics.csv!")
```
