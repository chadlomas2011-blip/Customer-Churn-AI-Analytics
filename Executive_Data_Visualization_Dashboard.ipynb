```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'], 'Tech_Sales': [12000, 15000, 18000, 14000, 22000, 26000], 'Services_Sales': [8000, 9500, 11000, 10500, 13000, 15500], 'Tech_Profit': [3600, 4500, 5400, 4200, 6600, 7800], 'Services_Profit': [1600, 1900, 2200, 2100, 2600, 3100]}
df = pd.DataFrame(data)

plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].plot(df['Month'], df['Tech_Sales'], marker='o', color='#1f77b4', linewidth=2.5, label='Tech')
axes[0, 0].plot(df['Month'], df['Services_Sales'], marker='s', color='#ff7f0e', linewidth=2.5, label='Services')
axes[0, 0].set_title('Monthly Sales Trajectory', fontsize=12, fontweight='bold')
axes[0, 0].legend()

df.plot(x='Month', y=['Tech_Sales', 'Services_Sales'], kind='bar', ax=axes[0, 1], color=['#1f77b4', '#ff7f0e'])
axes[0, 1].set_title('Sales Comparison by Category', fontsize=12, fontweight='bold')

sns.heatmap(df[['Tech_Sales', 'Services_Sales', 'Tech_Profit', 'Services_Profit']].corr(), annot=True, cmap='Blues', ax=axes[1, 0], fmt='.2f')
axes[1, 0].set_title('Sales vs. Profit Correlation', fontsize=12, fontweight='bold')

df['Total_Profit'] = df['Tech_Profit'] + df['Services_Profit']
axes[1, 1].pie(df['Total_Profit'], labels=df['Month'], autopct='%1.1f%%', colors=sns.color_palette('Blues', 6))
axes[1, 1].set_title('Profit Contribution by Month', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig("executive_dashboard.png", dpi=300)
plt.show()

print("--- DASHBOARD GENERATION COMPLETE ---")
print("Dashboard saved locally as 'executive_dashboard.png'")
```
