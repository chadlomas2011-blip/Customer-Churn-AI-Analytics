```python
import pandas as pd
import numpy as np

np.random.seed(42)
n_customers = 1000

ids = [f'CUST-{i:04d}' for i in range(1, n_customers + 1)]
contracts = np.random.choice(['Month-to-month', 'One year', 'Two year'], size=n_customers, p=[0.55, 0.25, 0.20])
charges = np.random.normal(loc=65, scale=20, size=n_customers).round(2)
tenures = np.random.randint(1, 72, size=n_customers)
churns = np.random.choice(['Yes', 'No'], size=n_customers, p=[0.27, 0.73])

df = pd.DataFrame({'customerID': ids, 'Contract': contracts, 'MonthlyCharges': charges, 'TenureMonths': tenures, 'Churn': churns})

total_customers = len(df)
churned = df[df['Churn'] == 'Yes']
churn_rate = (len(churned) / total_customers) * 100
lost_revenue = churned['MonthlyCharges'].sum()

print("--- BUSINESS METRICS SUMMARY ---")
print(f"Total Customers Analyzed: {total_customers:,}")
print(f"Overall Churn Rate: {churn_rate:.2f}%")
print(f"Monthly Revenue Lost: ${lost_revenue:,.2f}")
```

```python
# Sample customer feedback comments
comment1 = "My monthly bill jumped by $20 with no explanation!"
comment2 = "Wi-Fi connection keeps dropping every evening around 8 PM."
comment3 = "Waited on hold for 40 minutes and the representative hung up."
comment4 = "Internet speeds are super slow during storm days."
comment5 = "Double charged on my credit card this month."

feedback_list = [comment1, comment2, comment3, comment4, comment5]

# Zero-indent categorization using list comprehension
get_cat = lambda text: 'Billing & Pricing' if ('bill' in text.lower() or 'charged' in text.lower() or '$' in text.lower()) else ('Technical Issues' if ('wi-fi' in text.lower() or 'speed' in text.lower() or 'dropping' in text.lower()) else 'Customer Support')

categories = [get_cat(c) for c in feedback_list]

print("--- AI SENTIMENT CLASSIFICATION ---")
print(f"[{categories[0]}] --> '{comment1}'")
print(f"[{categories[1]}] --> '{comment2}'")
print(f"[{categories[2]}] --> '{comment3}'")
print(f"[{categories[3]}] --> '{comment4}'")
print(f"[{categories[4]}] --> '{comment5}'")
```
