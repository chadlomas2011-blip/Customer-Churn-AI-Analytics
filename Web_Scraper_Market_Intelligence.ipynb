```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

url = "http://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

titles = [a['title'] for a in soup.select("h3 a")]

raw_prices = [p.text for p in soup.select(".price_color")]
clean_prices = [float(re.findall(r"\d+\.\d+", p)[0]) for p in raw_prices]

availabilities = [a.text.strip() for a in soup.select(".availability")]

df = pd.DataFrame({"Item_Title": titles, "Price": clean_prices, "Availability": availabilities})

avg_price = df["Price"].mean()
max_price = df["Price"].max()
min_price = df["Price"].min()
total_items = len(df)

print("--- MARKET INTELLIGENCE SUMMARY ---")
print(f"Total Products Scraped: {total_items}")
print(f"Average Price: ${avg_price:.2f}")
print(f"Highest Price: ${max_price:.2f}")
print(f"Lowest Price: ${min_price:.2f}")
print("\n--- FIRST 5 SCRAPED RECORDS ---")
print(df.head())
```

```python
# Export scraped dataset to CSV file
df.to_csv("market_intelligence_data.csv", index=False)
print("Data successfully exported to market_intelligence_data.csv!")
```
