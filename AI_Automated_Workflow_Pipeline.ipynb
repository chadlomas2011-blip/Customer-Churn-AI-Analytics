```python
!pip install -q transformers torch pandas

import pandas as pd
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

tickets = ["CRITICAL ISSUE! The payment portal is down and my card was charged twice!", "Hi, do you offer an annual discount instead of monthly billing?", "The mobile app crashes every time I try to upload a profile picture.", "Just wanted to say I love the new dashboard layout!"]

categories = ["Urgent Escalation", "Billing Inquiry", "Technical Issue", "General Feedback"]

raw_outputs = [classifier(t, categories) for t in tickets]

data = [{"Ticket": r["sequence"][:35] + "...", "Category": r["labels"][0], "Confidence": f"{round(r['scores'][0]*100, 1)}%"} for r in raw_outputs]

df = pd.DataFrame(data)

print("--- AI WORKFLOW OUTPUT ---")
print(df)
```

```python
df.to_csv("ai_workflow_classification_results.csv", index=False)
print("Results successfully exported to ai_workflow_classification_results.csv!")
```
